## Prepare dataset

```python
X = df.drop(['species'], axis=1)

X.head()
```

```python
Y = df.species

Y.head()
```

## Label Encoding

```python
Y, class_names = pd.factorize(df.species)

class_names
```

```python
Y[:10]
```

## Train & Test Split

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=20)
```

```python
x_train.shape, x_test.shape
```

```python
y_train.shape, y_test.shape
```

```python
(label, count) = np.unique(y_test, return_counts=True)
freq = np.asarray((label, count)).T

print(freq)
```

## Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier()

model.fit(x_train, y_train)
```

```python
model.feature_importances_
```

## Tree

```python
from sklearn.tree import plot_tree

data_feature_names = X.columns.tolist()  # Convert Index to list
class_names_list = class_names.tolist()  # Convert Index to list

plt.subplots(nrows=1, ncols=1, figsize=(7, 7), dpi=96)
plot_tree(
    model,
    filled=True, rounded=True,
    feature_names=data_feature_names,
    class_names=class_names_list
)
```

## Evaluate

```python
score = model.score(x_test, y_test)

print(f"Score = {score.round(3)}")
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix

y_pred = model.predict(x_test)

print(f"Score -> {score:.3f}")
print(f"Accuracy Score -> {accuracy_score(y_test, y_pred):.3f}")

print(classification_report(y_test, y_pred))

print(classification_report(y_test, y_pred, target_names=class_names_list))

cm = confusion_matrix(y_test, y_pred)
print(cm)
```

## Confusion Matrics Display

```python
from sklearn.metrics import ConfusionMatrixDisplay

plt.rcParams.update({'font.size':12})
ConfusionMatrixDisplay(cm, display_labels=class_names).plot()
plt.title("Confusion Matrix")
```

## SVM

```python
from sklearn.svm import SVC

# model = SVC(kernel='linear')
model = SVC(kernel='rbf') # non-linear

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=20)

model.fit(x_train, y_train)
```

```python
score = model.score(x_test, y_test)

print(f"Score = {score.round(3)}")
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix

y_pred = model.predict(x_test)

print(f"Score -> {score:.3f}")
print(f"Accuracy Score -> {accuracy_score(y_test, y_pred):.3f}")

print(classification_report(y_test, y_pred, target_names=class_names_list))

cm = confusion_matrix(y_test, y_pred)
print(cm)
```

```python
from sklearn.metrics import ConfusionMatrixDisplay

plt.rcParams.update({'font.size':12})
ConfusionMatrixDisplay(cm, display_labels=class_names).plot()
plt.title("Confusion Matrix")
```
